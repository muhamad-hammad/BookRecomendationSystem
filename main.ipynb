```python
ratings.rename(columns=(
    {'User-ID':'user_id', 'Book-Rating':'rating'}

), inplace=True)
```

```python
ratings.head()
```

```python
x = ratings['user_id'].value_counts() > 200
```

```python
x[x].shape
```

```python
y = x[x].index
```

```python
y
```

```python
ratings = ratings[ratings['user_id'].isin(y)]
```

```python
ratings.shape
```

```python
RatingsWithBooks = ratings.merge(books, on='ISBN')
```

```python
RatingsWithBooks.head()
```

```python
numberRating = RatingsWithBooks.groupby('title')['rating'].count().reset_index()
```

```python
numberRating.rename(columns = {'rating':'number_of_ratings'}, inplace=True)
```

```python
numberRating.head()
```

```python
finalRating = RatingsWithBooks.merge(numberRating, on='title')
```

```python
finalRating.head()
```

```python
finalRating = finalRating[finalRating['number_of_ratings'] >= 50]
```

```python
finalRating.shape
```

```python
finalRating.drop_duplicates(['user_id', 'title'], inplace=True)
```

```python
finalRating.shape
```

```python
finalRatingPivot = finalRating.pivot_table(columns='user_id', index='title', values='rating')
```

```python
bookPivot = finalRating.pivot_table(columns='user_id', index='title', values='rating')
```

```python
bookPivot.head()
```

```python
bookPivot.shape
```

```python
bookPivot.fillna(0, inplace=True)
```

```python
bookPivot.head()
```

```python
#Traing knn model
from sklearn.neighbors import NearestNeighbors
NNeighbors = NearestNeighbors(algorithm='brute')
```

```python
NNeighbors.fit(bookPivot)
```

```python
_ , suggestions = NNeighbors.kneighbors(bookPivot.iloc[237, :].values.reshape(1,-1), n_neighbors=6)
```

```python
suggestions
```

```python
for i in range(len(suggestions)):
    print(bookPivot.index[suggestions[i]])
```

```python
#Poster link
bookName = []
for bookid in suggestions:
    bookName.append(bookPivot.index[bookid])
```

```python
bookName[0]
```

```python
idIndex = []
for name in bookName[0]:
    id = np.where(finalRating['title'] == name)[0][0]
    idIndex.append(id)
```

```python
for idx in idIndex:
  url = finalRating.iloc[idx]['img_url']
  print(url)
```

```python
bookNames = list(bookPivot.index)
```

```python
import pickle
pickle.dump(NNeighbors, open('NNeighbors.pkl', 'wb'))
pickle.dump(bookNames, open('bookNames.pkl', 'wb'))
pickle.dump(finalRating, open('finalRating.pkl', 'wb'))
pickle.dump(bookPivot, open('bookPivot.pkl', 'wb'))
```
